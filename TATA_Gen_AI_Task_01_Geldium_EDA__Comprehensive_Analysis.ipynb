```python
import pandas as pd

# Upload the file in Colab first using the file upload icon or below line:
# from google.colab import files
# uploaded = files.upload()

# Load Excel file
file_path = '/content/Delinquency_prediction_dataset.xlsm'
xls = pd.ExcelFile(file_path)

# Check available sheet names
print("Available Sheets:", xls.sheet_names)

# Load the main dataset
df = xls.parse('Delinquency_prediction_dataset')

# Show basic info
print("\n--- Dataset Info ---")
print(df.info())

# Display first 5 rows
print("\n--- First 5 Records ---")
display(df.head())

# Check missing values
print("\n--- Missing Values Count ---")
print(df.isnull().sum())

# Check for unique values in categorical columns
categorical_cols = df.select_dtypes(include='object').columns
print("\n--- Unique Values in Categorical Columns ---")
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")

# Describe numerical columns
print("\n--- Statistical Summary of Numeric Columns ---")
display(df.describe())
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Median Imputation for Income
income_imputer = SimpleImputer(strategy='median')
df['Income'] = income_imputer.fit_transform(df[['Income']])

# Mean Imputation for Credit_Score
credit_imputer = SimpleImputer(strategy='mean')
df['Credit_Score'] = credit_imputer.fit_transform(df[['Credit_Score']])

# Predictive Imputation for Loan_Balance
loan_df = df[['Loan_Balance', 'Income', 'Debt_to_Income_Ratio']].copy()
known = loan_df[loan_df['Loan_Balance'].notnull()]
missing = loan_df[loan_df['Loan_Balance'].isnull()]

# Fit linear regression model
reg = LinearRegression()
reg.fit(known[['Income', 'Debt_to_Income_Ratio']], known['Loan_Balance'])

# Predict missing values
predicted_loans = reg.predict(missing[['Income', 'Debt_to_Income_Ratio']])
df.loc[df['Loan_Balance'].isnull(), 'Loan_Balance'] = predicted_loans
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix (numeric features only)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation with Delinquent_Account")
plt.show()

# Boxplots for key numeric features
key_vars = ['Missed_Payments', 'Credit_Utilization', 'Debt_to_Income_Ratio']
for col in key_vars:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='Delinquent_Account', y=col, data=df)
    plt.title(f"{col} vs Delinquent_Account")
    plt.show()

# Bar plot for Employment_Status
plt.figure(figsize=(8, 4))
sns.countplot(x='Employment_Status', hue='Delinquent_Account', data=df)
plt.title("Employment Status vs Delinquency")
plt.xticks(rotation=45)
plt.show()
```
